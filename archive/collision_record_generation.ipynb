```python
import pandas as pd
import yaml #pip install pyyaml
import os
```

## Import tables

Downloaded from the HGNC custom downloads

```python
hgnc_custom_dl_df = pd.read_csv("downloaded_files/hgnc_custom_dl_20240828.txt", index_col= [0], sep="\t")
hgnc_custom_dl_df.rename(columns = {"Ensembl gene ID":"ENSG_ID"}, inplace = True)
hgnc_custom_dl_df.rename(columns = {"Approved symbol":"gene_symbol"}, inplace = True)
hgnc_custom_dl_df.rename(columns = {"UniProt ID(supplied by UniProt)":"UniProt_ID"}, inplace = True)
hgnc_custom_dl_df.rename(columns = {"OMIM ID(supplied by OMIM)":"OMIM_ID"}, inplace = True)
hgnc_custom_dl_df
```

This was created in the alias-alias collision nb

```python
merged_aa_collision_gene_df = pd.read_csv("created_files/merged_aa_collision_gene_df.csv", index_col= [0])
merged_aa_collision_gene_df
```

## Merge tables

Merging the two tables on ENSG ID and gene symbol

```python
suplemented_merged_aa_collision_gene_df = pd.merge(merged_aa_collision_gene_df, hgnc_custom_dl_df, how="left", on=["ENSG_ID","gene_symbol"])
suplemented_merged_aa_collision_gene_df
```

## Modify table

Added blank columns to provide fields for downstream manual curation

```python
suplemented_merged_aa_collision_gene_df["collision_gene_relationship"] = ""
suplemented_merged_aa_collision_gene_df["collision_acronym_expansion"] = ""
suplemented_merged_aa_collision_gene_df["collision_association"] = ""
suplemented_merged_aa_collision_gene_df["collision_source"] = ""
```

Remove white space from around alias symbols

```python
suplemented_merged_aa_collision_gene_df["collision"] = suplemented_merged_aa_collision_gene_df["collision"].apply(lambda x: x.strip() if isinstance(x, str) else x)
```

## Create a yaml file for each alias-alias collision symbol

If rerunning this cell, delete "aa yaml_files" folder first

```python
folder_path = "aa yaml_files"
os.makedirs(folder_path, exist_ok=True)
```

Group by alias symbol and gene symbol

```python
grouped = suplemented_merged_aa_collision_gene_df.groupby(["collision", "gene_symbol"])
```

Make a dictionary for yamls

```python
yaml_data = {}
```

```python
for (primary_val, secondary_val), group in grouped:
     # If the alias symbol is not already in the dictionary, add it   
    if primary_val not in yaml_data:
        yaml_data[primary_val] = {"collision_group": []}
        
    # Add gene symbols and associated data
    secondary_entry = {}
    secondary_entry["gene_symbol"] = secondary_val
    for col in ["ENSG_ID", "source", "Chromosome", "NCBI Gene ID", "OMIM_ID", "UniProt_ID"]:
        secondary_entry[col] = group[col].drop_duplicates().tolist()
    
    # Add collision gene relationship fields
    extra_field = {}
    extra_field["collision_acronym_expansion"] = group["collision_acronym_expansion"].drop_duplicates().tolist()
    extra_field["collision_association"] = group["collision_association"].drop_duplicates().tolist()
    extra_field["collision_source"] = group["collision_source"].drop_duplicates().tolist()
    
    
    secondary_entry["collision_gene_relationship"] = extra_field

    yaml_data[primary_val]["collision_group"].append(secondary_entry)
```

```python
additional_fields = {
    "collision_class": "", 
    "collision_type": ""        
}
```

Write each primary column value to a separate YAML file

```python
for primary_val, data_dict in yaml_data.items():
    data_with_key = {}
    data_with_key = {
        "collision_symbol": primary_val
    }
    data_with_key.update(additional_fields)

    data_with_key.update(data_dict)

    file_path = os.path.join(folder_path, f"{(primary_val.replace("/", "_"))}_collision_record.yaml")

    with open(file_path, "w") as wf:
        yaml.dump(data_with_key, wf, default_flow_style=None, sort_keys=False)

print("YAML files have been created.")
```
