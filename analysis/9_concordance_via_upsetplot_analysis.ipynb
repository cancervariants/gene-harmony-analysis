```python
def assign_prefix(value):
    if value.startswith('ENSG'):
        return "ENSG"
    elif value.startswith('HGNC'):
        return "HGNC"
    elif value.startswith('GENE'):
        return "GENE ID"
    else:
        return None
    
xref_freq_upset_df["prefix"] = xref_freq_upset_df['concept_id'].apply(assign_prefix)
xref_freq_upset_df
```

how often each combination of sources captured appears

```python
combination_counts = xref_freq_upset_df[source_list].agg(tuple, axis=1).value_counts()
```

```python
combination_counts
```

```python
xref_freq_upset_df
```

```python
combination_counts
```

converting the index of combination_counts from tuples to a hierarchical MultiIndex

```python
combination_counts.index = pd.MultiIndex.from_tuples(combination_counts.index)
```

```python
new_index = []

for level in combination_counts.index:
    # replace NaN/None values in each level of the MultiIndex tuple
    new_index.append(tuple(x if x is not None and pd.notna(x) else -1 for x in level))

combination_counts.index = pd.MultiIndex.from_tuples(new_index, names=combination_counts.index.names)
```

set the names of the levels in the MultiIndex of combination_counts to match the sources in source_list

```python
combination_counts.index.names = source_list
combination_counts.index.names
```
